```python
import numpy as np
from numpy.polynomial import chebyshev 
from matplotlib import pyplot as plt 
import math
```

```python
def polynomial_fit(x,y,order):
    """
    Function that fits the x and y data with a polynomial of order 
    "order". 
    Input(s): x: x-data (usually a range of numbers)
            y: y-data, function that is fitted
            order: Highest order of polynomial used to fit 
    Output(s): fit: the fit from polyfit and polyval
    """
    fit_poly = np.polyfit(x,y,order-1)
    fit = np.polyval(fit_poly,x)
    return fit

def chebyshev_mat(x,npoly):
    """
    Function that returns a matrix with each column being a chebyshev reccurence 
    polynomial. 
    Input(s): x: x-data (usually a range of numbers)
              npoly: An initial highest order for a polynomial
    Ouput(s): Matrix A with each column being a chebyshev polynomial. 
              The key relationship here is T_{n+1} = 2xT_{n}-T_{n-1}.
    """
    mat = np.zeros([x.size,npoly])
    mat[:,0] = 1.0 #T_0
    if npoly>1:
        mat[:,1] = x #T_1
    for i in range(2,npoly-1):
        mat[:,i] = 2.0*x*mat[:,i-1] - mat[:,i-2]
    mat = np.matrix(mat)
    return mat

def cheby_fitting(A,x,y,order):
    """
    Least Square fit to find the optimal coefficients using Chebyshev 
    polyonimals. 
    Input(s): x : x-range of the data
              y : function to be fitted (y-data)
              order: maximum order of chebyshev to fit with
    Output(s): Optimal coefficients for the fit of the Chebyshev polynomials
    """
    d = np.matrix(y).transpose()
    u,s,vt = np.linalg.svd(A,False)
    sinv = np.matrix(np.diag(1.0/s))
    fitp = vt.transpose()*sinv*(u.transpose()*d)
    return fitp
  
def coeff_tol(pp,tol):
    """
    Function that optimizes the order of the chebyshev polynomials. 
    Input(s): pp: coefficients of the Chebyshev polynomials
              tol: tolerance that we assign to the problem. In this 
                   question, it is 1e-6
    Output(s): Highest order of the polynomials which minimizes the error
               with respect to the tolerance. 
    """
    order = len(pp)
    for i in range(len(pp)):
        if np.abs(pp[i]) <= tol:
            order = i
            break
    try:
        return order
    except:
        print ("Tolerance cannot be reached in this order.")
        return len(pp)

def map_x(x):
    """
    Maps x into a range of [-1,1] using frexp() function. This function 
    seperates a value x into a pair (m,e) where x = m*2**e (e is an integer and 
    m is a float). m will be therefore a float that is lower than 1. 
    """
    fr = [math.frexp(t) for t in x]
    return np.asarray(fr)

#Initializing the true values of x and y   
x_org = np.arange(0.5,15,1e-3)
y_org = np.log2(x_org)


#Using frexp function to breakdown the x into m*2**e and hence
# y = log2(m) + e. So we just fit for m since that is bounded 
#between -1 and 1 and just add e.
fr = map_x(x_org)
x = fr[:,0]          #m
y = np.log2(x)       #log2(m)

#An initial order 
order = 100
   
#Chebyshev stuff
A = chebyshev_mat(x,order)         #Matrix A of polynomials
pp = cheby_fitting(A,x,y,order)    #Returned the optimized coefficients
order = coeff_tol(pp,1e-6)         #Truncated order when trying to find the optimized order of the polynomial
func = np.dot(A[:,:order],pp[:order])
func = np.squeeze(np.asarray(func))
func = func + fr[:,1]

poly = polynomial_fit(x_org,y_org,order)

#Plotting stuff
fig, [ax,ax1] = plt.subplots(2,1,gridspec_kw={'height_ratios':[1,0.3]},figsize=(10,6))
ax.plot(x_org,y_org,'.',color='black',label='True Values')
ax.plot(x_org,func,'--',color='red',label='fit with Chebyshev',linewidth=3)
ax.set_xlabel("x",fontsize=15)
ax.set_ylabel(r"$\log_2(x)$",fontsize=15)
ax.tick_params(labelsize=15)

func = np.squeeze(np.asarray(func))

ax1.plot(x_org,y_org-func,'.',color='black',label='Residuals')
print (f"Order of Chebyshev need to find the max error of {max(y_org-func)} is {order}. The RMS value is {np.sqrt(np.mean((y_org-func)**2))}")
print (f"The error from the Polynomial is {max(y_org-poly)}. The RMS value of that fit is {np.sqrt(np.mean((y_org-poly)**2))}")
ax.legend(fontsize=15)
plt.savefig('Chebyshev_fit.png')
plt.show()


fig, [ax,ax1] = plt.subplots(2,1,gridspec_kw={'height_ratios':[1,0.3]},figsize=(10,6))
ax.plot(x_org,y_org,'.',color='black',label='True Values')
ax.plot(x_org,polynomial_fit(x_org,y_org,order),'--',color='red',label='fit with Polynomials',linewidth=3)
ax.set_xlabel("x",fontsize=15)
ax.set_ylabel(r"$\log_2(x)$",fontsize=15)
ax.tick_params(labelsize=15)

ax1.plot(x_org,y_org-polynomial_fit(x_org,y_org,order),'.',color='black',label='Residuals')
ax.legend(fontsize=15)
plt.savefig('Polynomial_fit.png')
plt.show()
```

In the end, if we initialize the chebyshev polynomials with 50 terms, we only end up needing 48 terms. If we use 100 initial terms, we only need 98. Looking at the print statement above, the maximum error of the chebyshev fit is much lower than the polynomial fit. Looking at the residuals for the polynomial fit, this is due to the "ringing" that happens in the beginning and end, which does not occur in the chebyshev fit. The RMS value of the chebyshev fit is also lower than that of the polynomial fit, which means that the fit does not oscillate as much with cheby than it does with poly.
