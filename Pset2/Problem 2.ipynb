```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
time, flux, ignore = np.loadtxt('229614158_PDCSAP_SC6.txt',delimiter=',',unpack=True)
fig, ax = plt.subplots(1,1,figsize=(10,6))
plt.plot(time,flux)
```

```python
def calc_expdecay(guess,x):
    y = np.e**(-guess[0]*(x-guess[1]))+guess[2]
    grad = np.zeros([x.size,guess.size])
    grad[:,0] = -(x-guess[1])*np.e**(-guess[0]*(x-guess[1]))
    grad[:,1] = guess[0]*np.e**(-guess[0]*(x-guess[1]))
    grad[:,2] = 1
    return y, grad

def detect_Peak(x,y):
    """
    Function that returns the x value of the peak given y data
    Input(s): x : (float array) x-data (time)
              y : (float array) y-data (flux)
    Output(s): x value of the peak
    """
    y_max = max(y)
    ind = list(y).index(y_max)
    return (x[ind])

def exp(x,parms):
    return np.e**(-parms[0]*(x-parms[1]))+parms[2]
```

```python
#Prior guesses of e^(-B(x-C))+D
D = 1.0
B = 1/0.019
C = detect_Peak(time,flux) + np.log(max(flux)-D)/B

print ("The model would like the following: f(x) = e^(-B(x-C))+D")
print (f"The value of D is just an approximation of the steady state of the flux, which is {D}")
print (f"The value of B was found through trial and error - equal to {B}")
print (f"The value of C is found by using C = x+ln(a)*B where a is the amplitude of the flare, x is the beginning of the flare and B is the rate. This is equal to {C}")
```

The model chosen above is clearly not linear since we are using an exponential function to fit the flare. This model along with its guess parameters can be found above in the print statement. In summary, we use 

$$\begin{equation}
Y(x) = e^{B(x-C)}+D
\end{equation}$$

where D was just approximate to be on average just 1 (by looking at the time vs flux data), B was equal to 52.6 through trial and error and C was isolated using the coordinates of the amplitude of the flare and was found to be equal to around 1706.5. This model can be found below.

```python
fig,ax = plt.subplots(1,1,figsize=(10,6))
x = np.linspace(detect_Peak(time,flux),1710,531)
ind = list(time).index(detect_Peak(time,flux))
flux_cut = flux[ind:]
y = np.e**(-B*(x-C))+D
ax.plot(time,flux,'.',label='Raw Data of M-Dwarf Star',color='black')
ax.plot(x,y,'--',label='Guess parameters for the model',color='red')
ax.set_xlabel("time (s)",fontsize=13)
ax.set_ylabel(r"Flux $(W/m^2)$",fontsize=13)
ax.set_xlim(1706.4,1707)
ax.set_ylim(0.95,1.30)
ax.tick_params(labelsize=13)
ax.legend(fontsize=13)
plt.savefig('Model_approx.png')
plt.show()
```

Defining Newton's Method for part b)

```python
#Defining Newton's method way for part b) 

def NewtonMethod(x,y,p):
    """
    Implements Newton's method to minimize chi squared and hence 
    finding the best parameters for our functions. This will return 
    the parameters if the error on fit is less than 10% of its previous 
    iteration value 
    Input(s): x (float array): x-data (in this case, time)
              y (float array): y-data (in this case, flux)
              p (float array): guess parameters in the order [B,C,D]
    Output(s): p (float array): best fit parameters
               error (float array): chi squared of the fit
    """
    error = 1000
    for j in range(5):
        error_old = error
        pr, g = calc_expdecay(p,x)
        r = y - pr
        error = (r**2).sum()
        r = np.matrix(r).transpose()
        g = np.matrix(g)

        lhs = g.transpose()*g
        rhs = g.transpose()*r

        dp = np.linalg.inv(lhs)*rhs
        for k in range(p.size):
            p[k] = p[k] + dp[k]
        if error <1 and error - 0.1*error_old < 0 and j>0:
            break
        else:
            error = 1000  #Defining some random error if the error does not converge
    return p,error
```

```python
p = np.array([B,C,D])   #Intial guesses found in part a)
guess_P, chi_squared = NewtonMethod(x,flux_cut,p)


#Plotting stuff
fig, [ax,ax1,ax2] = plt.subplots(3,1,gridspec_kw={'height_ratios':[1,0.3,0.3]},figsize=(10,6))

y_pred = exp(x,guess_P)
ax.plot(x,y_pred,'-',label='Fit using Newton\'s Method',color='red')
ax.plot(time,flux,'.',label='Raw data for M-dwarf star',color='black')
ax.plot(x,y,'--',label='Fit with Best parameters from a)',color='purple')
ax.set_xlim(1706.4,1707)
ax.set_ylim(0.95,1.30)
ax.legend(fontsize=13)

#Residuals: 
ax1.plot(x,flux_cut-y_pred,'.',color='red')
ax2.plot(x,flux_cut-y,'.',color='purple')
ax2.set_xlabel("time (s)")
ax.set_ylabel(r"Flux $(W/m^2)$")
plt.savefig('Newton\'s_Method.png')

print (f"The best fit parameters are B = {guess_P[0]}, C = {guess_P[1]} and D = {guess_P[2]}")
print (f"The fit has a chi-squared of {chi_squared}")
```

```python
def error(x,y,guess):
    """
    Function that calculates the error of the parameters by looking at the noise
    of the data. 
    Input(s): x: (array floats) x-data (in this specific case, time)
              y: (array floats) y-data (in this specific case, flux)
              guess: Guessing parameters for the fit
    Output(s): err: error on each parameter 
               cov: covariance matrix
    """
    p, chi = NewtonMethod(x,y,guess)
    y_guess, grad = calc_expdecay(guess,x)
    noise = np.diag(1/((y-y_guess)*(y-y_guess)))
    cov = np.dot(grad.transpose(),noise)
    cov = np.dot(cov,grad)
    cov = np.linalg.inv(cov)
    err = np.sqrt(np.diag(cov))
    return err,cov
```

```python
err, cov = error(x,flux_cut,p)

print (f"The error on the B is {err[0]}. On C, it's {err[1]} and on D, it's {err[2]}")
print (f"The final fit parameters are B = {guess_P[0]} pm {err[0]}, C = {guess_P[1]} pm {err[1]}, D = {guess_P[2]} pm {err[2]}")


print ("I would not trust the errors on these data points since there are some other peaks \
we could be looking at. Morever, there seem to be correlated noise in the data judging by the residuals graphs.")
```

The error was calculated by looking at the noise of the data (aka difference between the data and the fit). From this, we can extract the covariance matrix and take the square root to find the error. The results were, 

$B = 66.70 \pm 0.08, C = 1706.5 \pm 6.2\times 10^{-5} and D = 0.99 \pm 1.1\times 10^{-5}$

These error seem resonnable. However, I feel like the error on C is a little bit too small for how the fit looks like. For example, the fit does not model correcly the part where the exponential start to stabilize; it seems that the data oscillates around the fit there. Therefore, I expected a higher error.
